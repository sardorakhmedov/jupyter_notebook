```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Первая функция (квадратичная)
def quadratic_function(x):
    return x**2 - x + 1

# График первой функции
x = np.linspace(-5, 5, 400)
y = quadratic_function(x)
plt.figure(figsize=(5, 5))
plt.subplot(2, 1, 1)
plt.plot(x, y, label=r'$y = x^2 - x + 1$', color='black')
plt.legend()

# Вторая функция (похожа на гармоническую, но измененная)
def modified_garmonic_function(x):
    return 2 * np.cos(2 * x + 1)  # Изменена синусоидальная на косинусоидальную

# График второй функции
x = np.linspace(-5, 5, 400)
y = modified_garmonic_function(x)
plt.subplot(2, 1, 2)
plt.plot(x, y, label=r'$y = 2 \cdot cos(2x + 1)$', color='black')

# Оформление
plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)
plt.legend(loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Первая функция (квадратичная)
def quadratic_function(x):
    return x**2 - x + 1

# Вторая функция (заменена на косинус)
def modified_function(x):
    return 2 * np.cos(2 * x + 1)  # Косинус вместо синуса

# Данные для графиков
x1 = np.linspace(-5, 5, 400)
x2 = np.linspace(-5, 5, 400)

y1 = quadratic_function(x1)
y2 = modified_function(x2)

# Построение графиков
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, label=r'$y = x^2 - x + 1$', color='black')  # Квадратичная функция
plt.plot(x2, y2, label=r'$y = 2 \cdot cos(2x + 1)$', color='red')  # Косинусоидальная функция
plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)
plt.legend(loc='upper left', fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Новый параметр t
t = np.linspace(0, 4 * np.pi, 400)

# Изменённые функции для x(t) и y(t)
x = np.cos(2 * t)  # Заменил на cos(2t)
y = np.sin(3 * t)  # Заменил на sin(3t)

# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=r'$x(t) = \cos(2t), \, y(t) = \sin(3t)$', color='black')
plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)
plt.legend(loc='upper left', fontsize=10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Годы
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

# Новая функция для прожиточного минимума (экспоненциальный рост)
a = 4340  # начальное значение прожиточного минимума
b = 0.05  # коэффициент роста

# Моделируем прожиточный минимум по экспоненциальной функции
minimum = a * np.exp(b * (np.array(years) - 2012))

# Построение графика
plt.figure(figsize=(10, 5))
plt.bar(years, minimum, color='black')
plt.xlabel('Год', fontsize=14)
plt.ylabel('Прожиточный минимум', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Годы
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

# Новая функция для прожиточного минимума (линейный рост, аналогичный)
# Например, начальный минимум и рост на 100 единиц каждый год
minimum = [4300, 4400, 4500, 4600, 4700, 4800, 4900, 5000, 5100, 5200, 5300]

# Построение круговой диаграммы
plt.figure(figsize=(10, 5))
plt.pie(minimum, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Распределение прожиточного минимума по годам (лінійний рост)', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание сетки
X, Y = np.mgrid[-5:5:10j, -5:5:10j]

# Новая функция для Z
Z = np.cos(X**2 + Y**2)  # Заменил sin(X^2) на cos(X^2 + Y^2)

# Создание 3D-оси
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Построение каркасной сетки
ax.plot_wireframe(X, Y, Z, color='blue')

# Показать график
plt.show()
```
